```python
# before using train test split we have to do encoding of categorical data 
```

```python
X = data.drop("species" , axis = 1)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
y = data["species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
# creating decision tree model with defult parameters
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
print(classification_report(y_test , y_pred))
```

```python
plot_confusion_matrix(model , X_test , y_test, cmap = "crest");
```

Lets create a function that reports the decision tree plot for different hyper parameter...

```python
def fun_report_model(model):
    
    y_pred_new = model.predict(X_test)
    
    print(classification_report(y_test , y_pred_new))
    
    plt.figure(dpi = 200)

    plot_tree(model , feature_names= X.columns , filled= True); 
    
```

```python
pruned_model = DecisionTreeClassifier(max_depth= 2) #max_depth = 2
```

```python
pruned_model.fit(X_train , y_train)
```

```python
fun_report_model(pruned_model)
```

```python
pruned_model = DecisionTreeClassifier(max_depth= 3) #max_depth = 3
```

```python
pruned_model.fit(X_train , y_train)
```

```python
fun_report_model(pruned_model)
```

```python
### Now lets create model with max_leaf_, earlier we used max_depth
```

```python
max_leaf_model = DecisionTreeClassifier(max_leaf_nodes= 3)
```

```python
max_leaf_model.fit(X_train , y_train)
```

```python
fun_report_model(max_leaf_model)
```

```python
### Lets now change impurity from "gini" to "entropy"
```

```python
entropy_model = DecisionTreeClassifier(criterion="entropy")
```

```python
entropy_model.fit(X_train , y_train)
```

```python
fun_report_model(entropy_model)
```
